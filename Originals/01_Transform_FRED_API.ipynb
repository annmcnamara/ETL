```python
### ETL TEAM 

#Extract year and date and concatenate both columns together as one column "month and year"
data_df["year"]  = pd.DatetimeIndex(data_df["Date"]).year
data_df["month"] = pd.DatetimeIndex(data_df["Date"]).month
data_df          = data_df.astype({"year": str, "month": str})

# We need monthyear to search the other data
data_df["month&year"] = data_df["year"] + data_df["month"]
data_df.head()
```

```python
#Group the dataframe by month year
grouped = data_df.groupby("month&year")
```

```python
#Calculate mean of the index in group
## Use numpy to aggregate mean
aggregated_series = grouped[["DCOILBRENTEU", "DJIA", "DCOILWTICO","DHHNGSP","GOLDAMGBD228NLBM","NASDAQCOM","SP500","WILL5000INDFC","DEXCHUS","DEXINUS","DEXJPUS","DEXUSEU"]].agg(np.mean)
aggregated_df = pd.DataFrame(aggregated_series)
```
